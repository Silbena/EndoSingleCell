```python
import scanpy as sc
from pathlib import Path
```

```python
sc.settings.set_figure_params(dpi=100, facecolor="white")
```

```python
data_file = Path('../data/endometrium_filtered.h5ad')
data = sc.read_h5ad(data_file)
```

# Feature selection

```python
sc.pp.highly_variable_genes(data, n_top_genes=2000, batch_key='SampleID')
```

```python
sc.pl.highly_variable_genes(data, save='features.png')
```

# PCA

```python
sc.tl.pca(data, n_comps=50)
```

Elbow plot

```python
sc.pl.pca_variance_ratio(data,
                         n_pcs=30,
                         log=True,
                         save='elbow.png')
```

Based on the elbow plot I will continue with 20 PCs.

```python
sc.pl.pca(data,
          color=['Broad cell type'],
          dimensions=[(0, 1)],
          ncols=1,
          size=1)
```
